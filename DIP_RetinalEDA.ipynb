Listing the Files

```python
import os
# List the files in the folder
image_folder = 'sample'
image_files = os.listdir(image_folder)

print(image_files)  # Prints the list of image filenames
```

Loading Images Using OpenCV

```python
import cv2
import matplotlib.pyplot as plt

for image_file in image_files:
    img_path = os.path.join(image_folder, image_file)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB for displaying
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
print(f'Shape: {img.shape}')  # (height, width, channels)
```

Histogram


```python
for i, col in enumerate(['r', 'g', 'b']):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)
    plt.xlim([0, 256])
plt.show()
#rgb graph
```

Pixel Intensity

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.hist(gray_img.ravel(), 256, [0, 256])
plt.show()
```

Edge Detection Using Canny


```python
edges = cv2.Canny(gray_img, 100, 200)
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.show()
```

Thresholding

```python
_, thresh = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresh, cmap='gray')
plt.axis('off')
plt.show()
```

Noise Analysis

```python
from skimage.util import random_noise
noisy_image = random_noise(img, mode='gaussian')
plt.imshow(noisy_image)
plt.axis('off')
plt.show()
```

Resizing Images

```python
resized_img = cv2.resize(img, (128, 128))  # Resize to 128x128
```

Normalizing Images

```python
normalized_img = img / 255.0
```

Noise Removal

```python
denoised_img = cv2.GaussianBlur(img, (5, 5), 0)
```

Displaying Images After Preprocessing

```python
# Plot original and processed images side by side

plt.figure(figsize=(12, 12))

# Original image
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Resized image
plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))
plt.title('Resized Image (128x128)')
plt.axis('off')

# Grayscale image
plt.subplot(2, 2, 3)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# Normalized image
plt.subplot(2, 2, 4)
plt.imshow(normalized_img)
plt.title('Normalized Image')
plt.axis('off')

plt.show()
```

```python
# Apply Canny edge detection
edges = cv2.Canny(gray_img, 100, 200)

# Display the edges
plt.figure(figsize=(6, 6))
plt.imshow(edges, cmap='gray')  # Display edges as a grayscale image
plt.title('Edge Detection')
plt.axis('off')
plt.show()
```
